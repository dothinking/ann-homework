real images with labels

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Check Original Image

```python
# read images path
import glob
samples = glob.glob('samples/*.jpg')
len(samples)
```

```python
# load image
img_raw = tf.io.read_file(samples[0])
img = tf.image.decode_image(img_raw)

print(img.shape) # (height, width, channel)
print(np.min(img.numpy()), np.max(img.numpy())) # int [0, 255]
```

```python
# show image
plt.imshow(img)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.xlabel(samples[0][-8:])
plt.show()
```

## 2. PreProcessing

- resize
- grayscale

```python
# convert to floats in the [0,1] range
img = tf.image.convert_image_dtype(img, tf.float32)
print(type(img))
np.min(img), np.max(img)
```

```python
# resize
img = tf.image.resize(img, [60, 120])
```

```python
# convert RGB to grayscale
img = tf.image.rgb_to_grayscale(img)
img.shape
```

```python
# show resized image
img = tf.reshape(img, (60, 120))
plt.imshow(img, cmap=plt.cm.binary)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.xlabel(samples[0][-8:])
plt.show()
```
